```python
import pandas as pd
df=pd.read_csv('D:/Jalsha deck/Web scraped data/scr/cleaned_code_SJ/diya-aur-baati-hum.csv',encoding='utf-8')

# create month and year variable 

from datetime import datetime
import calendar 
   
df['month']=df['Date'].apply(lambda x:str(x).strip('00:00').split('-')[1])
df['year']=df['Date'].apply(lambda x:str(x).strip('00:00').split('-')[2])
df['counter']=1

# no of episodes per month per year
a=df.groupby(['month','year'])['Date'].nunique()

# create a column no of words per scene level

def wds(x):
    return len((x.split(" ")))
df['No of words']=df['Para text'].apply(wds)

# no of words per episode per month per year
print (df.groupby(['month','year'])['No of words'].sum()/df.groupby(['month','year'])['Date'].nunique())

#  cleaning tokenising and extracting sentences per scene

import nltk
import re

#from nltk import * 

def fn_clean_para(a):
    a=re.sub(r'(mt.)','mt ',a,flags=re.I)
    a=re.sub(r'(st.)','st ',a,flags=re.I)
    a=re.sub(r'(mr.)','mr ',a,flags=re.I)
    a=re.sub(r'(mrs.)','mrs ',a,flags=re.I)
    a=re.sub(r'(\\xe2\\x80\\x99s)','',a)
    #a=re.sub(r'(\xe2\u20ac\xa6)','',a)
    #a=re.sub(r'(:::)','',a)
    return a
df['new_text']=df['Para text'].apply(fn_clean_para)
df['sentence_count']=df['new_text'].apply(lambda x:len(nltk.sent_tokenize(x)))

# get no of sentence per episode per month per year 
# sentence count per episode per month per year 

(df.groupby(['month','year'])['sentence_count'].sum()/df.groupby(['month','year'])['Date'].nunique())
```
